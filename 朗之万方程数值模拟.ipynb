$$
d x=-\frac{(x-\mu)}{\tau} d t+\sigma \sqrt{\frac{2}{\tau}} d W
$$

$$
d x=a(t, x) d t+b(t, x) d W
$$

$$
x_{n+1}=x_{n}+d x=x_{n}+a\left(t, x_{n}\right) d t+b\left(t, x_{n}\right) \sqrt{d t} \xi, \quad \xi \sim N(0,1)
$$

```python
sigma = 1.  # Standard deviation.
mu = 0.  # Mean.
tau = .05  # Time constant.

dt = .001  # Time step.
T = 1.  # Total time.
n = int(T / dt)  # Number of time steps.
t = np.linspace(0., T, n)  # Vector of times.

sigma_bis = sigma * np.sqrt(2. / tau)
sqrtdt = np.sqrt(dt)
```

```python
def create_trajectory(x0 = 0, n = 1000, sigma = 1, mu = 0, tau = .05):

    dt = 1/n   # dt越小，随机性越大吧？
    
    v = np.zeros(n)

    for i in range(n - 1):
        v[i + 1] = v[i] - dt * (v[i] - mu) / tau + \
            sigma * np.sqrt(2. / tau) * np.sqrt(dt) * np.random.normal(0, 1)
    
    x = x0 + np.cumsum(v) #不乘dt，相当于放大横纵坐标轴
    
    return x
```

```python
%matplotlib inline
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,6))
plt.title('sigma = 1, mu = 0, tau = .05')
x = create_trajectory(tau = .05)
y = create_trajectory(tau = .05)
plt.plot(x, y)
plt.show()
```

# 从速度的朗之万方程出发，改变参数

$$
\begin{aligned}
\frac{\mathrm{d} x(t)}{\mathrm{d} t} &=v(t) \\
\frac{\mathrm{d} v(t)}{\mathrm{d} t} &=-\frac{\gamma}{m} v(t)+\frac{1}{m} \xi(t)
\end{aligned}
$$

$$
\gamma=6 \pi \eta a 
$$

$$
\tau_{B} \approx \frac{m}{\gamma} \approx 10^{-3} \mathrm{~S}
$$

$$
v_{n+1}=v_{n}+dv=v_{n}-v_{n}\frac{dt}{\tau}+{\mu}\frac{dt}{\tau}+\sigma \sqrt{2\frac{dt}{\tau}} \xi, \quad \xi \sim N(0,1)
$$

```python
#t = np.linspace(0., T, n)  # Vector of times.
```

```python
# for paper

def trajectory_two(x0=[0,1],y0=[0,1],n = 1000, sigma = 1, tau = 0.05, r = [5,5], mu = 0.1 ):
    
    dt = 1/n
    
    r1 = r[0]
    r2 = r[1]
        
    tau1 = tau  # 严谨一点的话应该是r^2的比例
    tau2 = tau
       
    vx1 = np.zeros(n)
    vy1 = np.zeros(n)
    
    vx2 = np.zeros(n)
    vy2 = np.zeros(n)


    for i in range(n - 1):
        
        x1 = sum(vx1)+x0[0]
        y1 = sum(vy1)+y0[0]
        
        x2 = sum(vx2)+x0[1]
        y2 = sum(vy2)+y0[1]

        
        if (x1-x2)**2 + (y1-y2)**2 >= 2*(r1+r2)**2 :
        
            vx1[i + 1] = vx1[i] - dt * vx1[i] / tau1 +  np.sqrt(2. / tau1) * np.sqrt(dt) * np.random.normal(0, 1)
            vx2[i + 1] = vx2[i] - dt * vx2[i] / tau2 +  np.sqrt(2. / tau2) * np.sqrt(dt) * np.random.normal(0, 1)
            
            vy1[i + 1] = vy1[i] - dt * vy1[i] / tau1 +  np.sqrt(2. / tau1) * np.sqrt(dt) * np.random.normal(0, 1)
            vy2[i + 1] = vy2[i] - dt * vy2[i] / tau2 +  np.sqrt(2. / tau2) * np.sqrt(dt) * np.random.normal(0, 1)
        
        
        if (x1-x2)**2 + (y1-y2)**2 < 2*(r1+r2)**2 :
            
            vx1[i + 1] = vx1[i] - dt * (vx1[i]+mu*(vx1[i] - vx2[i])/(x1-x2+0.01)) / tau1  \
                         +  np.sqrt(2. / tau2) * np.sqrt(dt) * np.random.normal(0, 1)
                                        
            vx2[i + 1] = vx2[i] - dt * (vx2[i]+mu*(vx2[i] - vx1[i])/(x2-x1+0.01)) / tau2  \
                         +  np.sqrt(2. / tau2) * np.sqrt(dt) * np.random.normal(0, 1)
                
            vy1[i + 1] = vy1[i] - dt * (vy1[i]+mu*(vy1[i] - vy2[i])/(y1-y2+0.01)) / tau1 \
                         +  np.sqrt(2. / tau1) * np.sqrt(dt) * np.random.normal(0, 1)
            vy2[i + 1] = vy2[i] - dt * (vy2[i]+mu*(vy2[i] - vy1[i])/(y2-y1+0.01)) / tau2 \
                         +  np.sqrt(2. / tau2) * np.sqrt(dt) * np.random.normal(0, 1)          
               
    
    x1_tra = np.cumsum(vx1)+x0[0]
    y1_tra = np.cumsum(vy1)+y0[0]
    
    x2_tra = np.cumsum(vx2)+x0[1]
    y2_tra = np.cumsum(vy2)+y0[1]
    
    return x1_tra,x2_tra,y1_tra,y2_tra
```

```python
%matplotlib inline
x1_tra,x2_tra,y1_tra,y2_tra = trajectory_two(r = [5,5], mu = 100)
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.plot(x1_tra[:100], y1_tra[:100])
ax.plot(x2_tra[:100], y2_tra[:100])
```
